```python
path=r"C:\Users\ladom\Desktop\LLm\data\SeoulBikeData.csv"
cols=["bike_count", "hour", "temp", "humidity", "wind", "visibility", "dew_pt_temp" ,"radiation", "rain", "snow", "functional"]
df=pd.read_csv(path, encoding="cp949").drop(["Date", "Holiday", "Seasons"], axis=1)
df.columns=cols
df["functional"]=(df["functional"]=="Yes").astype(int)
df=df[df["hour"]==12]
df=df.drop(["hour"], axis=1)
df.head()
```

```python
for label in df.columns[1:]:
    plt.scatter(df[label], df["bike_count"])
    plt.title(label)
    plt.xlabel(label)
    plt.ylabel("bike count at noon")
    plt.show()
```
